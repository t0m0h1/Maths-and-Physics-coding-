#### Area and perimeter of a circle

```python
import math

# Area=Ï€r2

pi = math.pi

# Area
def area(radius):
    val = pi * radius * radius
    return round(val, 2)

print(area(6))
```

### Distance between two points (Euclidean Distance)
- Write a function that calculates the distance between two points 
- This is the same as drawing a line between point A and point B

```python
import numpy as np

point1 = np.array([1, 2, 3])
point2 = np.array([4, 6, 8])

def distance(p1, p2):
    return np.linalg.norm(p1 - p2)

dist = distance(point1, point2)
print("Euclidean Distance:", dist)
```

### Create 3D visualisation of Euclidean Distance

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Points
point1 = np.array([1, 2, 3])
point2 = np.array([4, 6, 8])

# Function to calculate distance
def distance(p1, p2):
    return np.linalg.norm(p1 - p2)

# Calculate distance
dist = distance(point1, point2)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(*point1, color='red', label='Point 1 (1, 2, 3)')
ax.scatter(*point2, color='blue', label='Point 2 (4, 6, 8)')

ax.plot([point1[0], point2[0]],
        [point1[1], point2[1]],
        [point1[2], point2[2]], color='green', label=f'Distance: {dist:.2f}')

# Set labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

ax.legend()
plt.show()
```
